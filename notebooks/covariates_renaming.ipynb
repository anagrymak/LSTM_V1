# Covariates renaming

This notebook does a simple convertion of format for covariates, from underscore separated names to multiple columns in a .csv file.

```python
import pandas as pd
```

```python
dset = pd.read_csv("../data/conventional_namings_raw.csv", header=None)
```

```python
dset = dset.rename(columns={0: "material", 1: "material_type"})
```

```python
dset
```

```python
dset = dset.join(dset.material_type.str.split("_", expand=True)).drop(columns="material_type")
dset
```

```python
dset[3] = dset[3].str[:-1].astype(int)
```

```python
dset.columns = ["material", "material_type", "medium", "glazed", "force_N"]
```

```python
dset
```

```python
dset.to_csv("covariates.csv", index=False)
```
